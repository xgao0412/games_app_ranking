```python
## news feed

import sys
r_stream = requests.get('https://stream.twitter.com/1.1/statuses/sample.json', auth=auth, stream=True)
counter = 0
for line in r_stream.iter_lines():
    # filter out keep-alive new lines
    if not line:
        continue
    tweet = json.loads(line)
    if 'text' in tweet:
        counter +=1
        print(tweet['text'])
    sys.stdout.flush()
    if counter > 100:
        break
```

```python
import datetime as dt

def get_tweet(app):
    return query_tweets(app, limit=None, 
                         begindate=dt.date.today()-dt.timedelta(days=7), 
                         enddate=dt.date.today(), 
                         poolsize=20, 
                         lang='en'
                         )
```

```python
import spacy
spacy.load('en')
from spacy.lang.en import English
parser = English()

def tokenize(text):
    lda_tokens = []
    tokens = parser(text)

    for token in tokens:
        if token.like_url:
            lda_tokens.append(token)
    return lda_tokens
```

```python
## hoteness and url
# 3 days
# limit output
from notebook.services.config import ConfigManager
cm = ConfigManager().update('notebook', {'limit_output':100,'limit_output_message':100,'limit_display_data':100,
                                         'limit_execute_result':100, 'limit_stream':100})


import itertools
from collections import Counter
import time

t1 = time.time()

# testing 
test_app = app_name

data_array=[]
for app in test_app:
    tweet = get_tweet(app);

    length = len(tweet)
    likes= 0
    reply= 0
    retweets= 0
    url = []
    
    if tweet:
        
        length = len(tweet)
        likes= 0
        reply= 0
        retweets= 0
        url = []
        
        for line in tweet:
            likes+=int(line.likes)
            reply+=int(line.replies)
            retweets+=int(line.retweets)
            
            if line.text:
                #print(line.text)
                c = tokenize(line.text)
                if c:
                    url.append(c)

        if url:
            se = list(itertools.chain(*url))
            topurl = Counter(se).most_common(1)[0][0]

        data_array.append([app,length,likes,reply,retweets,topurl])

        
    #else:data_array.append([])
    print(app)
print('running time',(time.time()-t1))    
```

```python
import pandas as pd
data_df = pd.DataFrame(data_array,columns=['app','No.of tweets','No.of likes','No.of replies','No.of retweets','most common url'])
```
